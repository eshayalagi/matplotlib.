üîπ What is a Stack Plot?
A stack plot (or stacked area chart) is a type of plot that shows multiple data series stacked on top of each other. It helps visualize how individual components contribute to the whole over time or across a sequence.

Each layer in a stack plot represents a category, and the cumulative height of the areas shows the total at each point.

üîπ When to Use Stack Plots in EDA?
Stack plots are useful for:

Showing changes over time for multiple groups

Analyzing category-wise trends that contribute to a total

Visualizing time-series categorical data

Understanding proportions and how they evolve

‚úÖ Best for continuous data
‚ùå Not ideal for raw frequency comparison (bar charts are better there)



```python
import matplotlib.pyplot as plt

activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3, 2, 2, 5]  # hours in a day

colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']

plt.figure(figsize=(6,6))
plt.pie(time_spent, labels=activities, autopct='%1.1f%%', startangle=90)
plt.title("After School Activities")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

What is a Stack Plot?

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']

plt.figure(figsize=(10,6))
plt.stackplot(days, studying, playing, watching_tv, sleeping,
              labels=labels, colors=colors, alpha=1)

plt.legend(loc='upper left')  # Location of the legend
plt.title('Weekly Activity Tracker')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.grid(True)
plt.show()
plt.grad()
```
