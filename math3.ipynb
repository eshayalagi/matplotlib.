🥧 Pie Charts in EDA using Matplotlib
🔹 What is a Pie Chart?
A pie chart is a circular statistical plot used to show the proportions or percentages of a whole. Each slice of the pie represents a category's contribution to the total.

🔹 When to Use Pie Charts in EDA?
Pie charts are useful for:

Showing the relative proportion of categories

Quickly understanding dominant categories

Comparing a small number of categories (ideally <6)

Visualizing class balance in datasets

✅ Best for high-level summaries
❌ Not ideal for detailed comparisons or large category counts

🔹 Matplotlib for Pie Charts
Matplotlib provides a simple method plt.pie() to draw pie charts.

🔸 Syntax:

plt.pie(values, labels=categories, autopct='%1.1f%%')

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")  # Choose any style you like
```

```python
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 950]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels,colors=colors)
plt.show()
```

https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Wedge.html

```python
 plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'red', 'linewidth': 1, 'linestyle': '-','fill':True}
)
plt.show()
```

Highlight a Slice Using explode
Use explode to pull out one or more slices.

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=colors,shadow=True,startangle=100, autopct='%3f%%',)
##(to rotate the trangle)
plt.title("Exploded Pie Example")
plt.show()
```

Crowded Pie Chart – Why to Avoid

```python
# Too many categories
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```
