Subplot in Data Science / Data Visualization
In data science, the term subplot is most commonly used in data visualization â€” especially with Python libraries like Matplotlib or Seaborn.

âœ… Definition:
A subplot refers to a single plot (chart/graph) that is part of a larger figure containing multiple plots.

ðŸ“Œ In simpler terms:

A subplot is one of many graphs in a single window or figure.

ðŸ§  Why use subplots?
To compare multiple variables or charts side-by-side

To show different views of the same data

To make visual analysis easier and cleaner



```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4]
y1=[i*2 for i in x]
y2=[i**2 for i in x]
y3=[i*5 for i in x]
y4=[i*10 for i in x]
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)  
plt.plot(x, y2)
plt.title('Square of x')
plt.xlabel('Number')
plt.ylabel('Square')

plt.subplot(2, 1, 2) 
plt.plot(x, y1)
plt.title('Double of x')
plt.xlabel('Number')
plt.ylabel('Double')

plt.tight_layout() 
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('x*2')


plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('x**2')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('i*5')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('i*10')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 4))

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('x * 2')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('x squared')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('x * 10')  # Corrected title

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('x * 5')   # Corrected title

fig.suptitle('Simple Comparison Plot', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.85)  # Adjusting title position
fig.savefig('my_plots.png')    # Save image

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y1)
axs[0].set_title('x * 2')

axs[1].plot(x, y2)
axs[1].set_title('x squared')
 
fig.suptitle('Simple Comparison Plots', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.85)  # So title doesn't overlap
fig.savefig('my_plots.png')    # Save as image

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
ys = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']

for i in range(2):
    for j in range(2):
        idx = i * 2 + j
        axs[i, j].plot(x, ys[idx])
        axs[i, j].set_title(titles[idx])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create 1 figure with 2 subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2)

# First subplot
axs[0].plot([1, 2, 3], [1, 4, 9])
axs[0].set_title('Plot 1')

# Second subplot
axs[1].plot([1, 2, 3], [9, 4, 1])
axs[1].set_title('Plot 2')

plt.show()
```
