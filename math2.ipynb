ðŸ”¹ What is a Bar Graph in EDA?
A bar graph (or bar chart) displays the frequency or count (or sometimes another statistic like mean) of categories as rectangular bars. The height (or length, in horizontal bar charts) of each bar represents the value associated with each category.

ðŸ”¹ When to Use Bar Graphs in EDA?
To compare the size of different categories

To detect imbalances or dominant categories

To find missing or rare values

To analyze class distribution in a classification problem



```python
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.bar(years,runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
import numpy as np

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width=0.1
```

Setting Bar Width & Side-by-Side Bar Charts

```python
plt.bar(x + width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x - width, kohli, width=width, label="Kohli")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
plt.show()
```

Horizontal Bar Charts with barh()  when categorical and numerical values should be 
shown(plt.bar()	Vertical bars for categorical comparison
plt.barh()	Horizontal bars (great for long labels)) 

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
```

```python
runs_5yrs
```

```python
plt.barh(players,runs_5yrs,color="red")
plt.xlable("total run in 1st 5 yrs")
plt.tittle("First 5-Year Performance of Indian Batsmen")
plt.show()
```

```python
import matplotlib.pyplot as plt

players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color="skyblue")

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 50, str(runs[i]), ha='center')

plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
plt.show()
```
